```python
import os
import numpy as np
import pandas as pd
!pip install pmaw
from pmaw import PushshiftAPI
from datetime import datetime

sub_r = 'TrueAskReddit'
begin = int(datetime(2010,12,30).timestamp())

path = '/content/drive/MyDrive/reddit_data/' + sub_r + '_submissions/'
if not os.path.exists(path):
    os.mkdir(path)
    
api = PushshiftAPI()
```

```python
# find last time point
def get_num(string):
    num = ''
    for c in string:
        if c.isdigit():
          num += c
    return int(num)

last_file = ''
last_num = -1

if not os.path.exists(path):
    os.mkdir(path)
    
if os.listdir(path):
    files = next(os.walk(path))[2]
    for f in files:
        num = get_num(f)
        if num > last_num:
            last_num = num
            last_file = f

if last_file:
    df = pd.read_csv(path + last_file, usecols=['created_utc'])
    index = [(isinstance(x, str) | isinstance(x, int)) for x in df['created_utc']]
    last_time_point = int(min(df['created_utc'][index]))
else:
    last_time_point = int(datetime.now().timestamp())

datetime.utcfromtimestamp(last_time_point), last_num
```

```python
end = last_time_point
time_list = range(end, begin, -60*60*24)

submissions  = []
sub_list = []
submissions_all  = []
count = last_num + 1

for t in time_list:
  try:
    submissions = api.search_submissions(subreddit=sub_r, before = t, after = t-60*60*24, limit = 3000)
    sub_list = [p for p in submissions]
    submissions_all.extend(sub_list)
  except:
    continue
  
  if len(submissions_all) > 2000:
    df = pd.DataFrame(submissions_all)
    df.to_csv(path + sub_r + '_submissions_{}.csv'.format(count))
    submissions_all = []
    count += 1

if submissions_all:
    df = pd.DataFrame(submissions_all)
    df.to_csv(path + sub_r + '_submissions_{}.csv'.format(count))
```
